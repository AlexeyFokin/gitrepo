```python
def sp(string):
  ar = string.str.split()
  nextstr = ''
  final = ''
  for word in ar:
    nextstr += word + ' '
    if len(nextstr) > 70:
      final += nextstr + '\n'
      nextstr = ''
    
  return(final)
```
