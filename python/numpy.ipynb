```
import numpy as np
def get_intersect(data):
    """
    Returns the point of intersection of the lines passing through a2,a1 and b2,b1.
    a1: [x, y] a point on the first line
    a2: [x, y] another point on the first line
    b1: [x, y] a point on the second line
    b2: [x, y] another point on the second line
    """
    a1, a2, b1, b2 = data
    s = np.vstack([a1,a2,b1,b2])        # s for stacked
    h = np.hstack((s, np.ones((4, 1)))) # h for homogeneous
    l1 = np.cross(h[0], h[1])           # get first line
    l2 = np.cross(h[2], h[3])           # get second line
    x, y, z = np.cross(l1, l2)          # point of intersection
    if z == 0:                          # lines are parallel
        return (float('inf'), float('inf'))
    return (x/z, y/z)



```

```
def get_intersect(twolines):
    """
    Returns the point of intersection of the lines passing through a2,a1 and b2,b1.
    a1: [x, y] a point on the first line
    a2: [x, y] another point on the first line
    b1: [x, y] a point on the second line
    b2: [x, y] another point on the second line
    """
    line1, line2 = twolines
    a1, a2 = line1.reshape(2,2)
    b1, b2 = line2.reshape(2, 2)
    #a1, a2 = line1
    #b1, b2 = line2

    s = np.vstack([a1, a2, b1, b2])        # s for stacked
    h = np.hstack((s, np.ones((4, 1)))) # h for homogeneous
    l1 = np.cross(h[0], h[1])           # get first line
    l2 = np.cross(h[2], h[3])           # get second line
    x, y, z = np.cross(l1, l2)          # point of intersection
    if z == 0:                          # lines are parallel
        return (float('inf'), float('inf'))
    return (x/z, y/z)
```

```
class Graph:

    # init function to declare class variables
    def __init__(self, V):
        self.V = V
        self.adj = {}

    # method to add an undirected edge
    def addEdge(self, v, w):
        try:
          a = self.adj[v]
          a = a.append(w)
          self.adj[v] = a
        except KeyError:
          self.adj[v] = (w)
        try:
          self.adj[w].append(v)
        except KeyError:
          self.adj[w] = (v)

    def DFSUtil(self, temp, v, visited):

        # Mark the current vertex as visited
        visited[v] = True

        # Store the vertex to list
        temp.append(v)

        # Repeat for all vertices adjacent
        # to this vertex v
        for i in self.adj[v]:
            if visited[i] == False:
                # Update the list
                temp = self.DFSUtil(temp, i, visited)
        return temp


    # Method to retrieve connected components
    # in an undirected graph
    def connectedComponents(self):
        visited = []
        cc = []
        for i in range(self.V):
            visited.append(False)
        for v in range(self.V):
            if visited[v] == False:
                temp = []
                cc.append(self.DFSUtil(temp, v, visited))
        return cc
```

```
# Тренировка пайтон-пандас.
# сгенерить функцию нормального распределения. и посчитать ее дисперсию с помощью пандас

# def normal_destr(sigma, mu):

import math
import pandas as pd
import numpy as np
from scipy.stats import norm
import plotly.express as px

class NormDestr:
  def __init__(self, mu, sigma):
    self.sigma = sigma
    self.mu = mu

  def norm(self, x):
    coef = 1 / (self.sigma * math.sqrt(2 * math.pi))
    exponent = -1/2 * (x - self.mu / self.sigma) ** 2
    return coef * math.exp(exponent)


nrm = NormDestr(0, 1)

x = np.arange(-4, 4, 0.00001)
y1 = [nrm.norm(a) for a in x]
y = norm.pdf(x, 0, 1)


ds = pd.Series(y1)
ds.mean()

fig = px.line(x=x, y=y, title='Normal distribution')
#fig.show()
fig1 = px.line(x=x, y=y1, title='Normal distribution')
#fig1.show()

print (ds.mean())
```
