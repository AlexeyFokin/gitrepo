```
import plotly.express as px 
import numpy as np
  
# creating random data through randomint  
# function of numpy.random  
np.random.seed(42)  
    
random_x= np.random.randint(1,101,100)  
random_y= np.random.randint(1,101,100) 
  
fig = px.bar(random_x, y = random_y) 
 
#fig.update_layout(
#    #autosize=False,
#    width=1000,
#    height=400,
#)
fig.show()
```
