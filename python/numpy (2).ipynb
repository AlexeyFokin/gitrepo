```
# Import numpy and matplotlib.pyplot
import numpy as np
import matplotlib.pyplot as plt

# Generate two 1-D arrays: u, v
u = np.linspace(30, 60, 20)
v = np.linspace(30, 60, 20)

# Generate 2-D arrays from u and v: X, Y
X,Y = np.meshgrid(u, v)

# Compute Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

# Display the resulting image with pcolor()
plt.pcolor(Z)
plt.show()

# Save the figure to 'sine_mesh.png'
#plt.savefig('sine_mesh.png')
```

```
start_time = time.clock()

# size of the box
#size = (1080, 1920)
#size = (500, 500)
size = (1440, 2560)

# step between concentric circles
step = 15

# radius of the point, 0 - single pixel
thickness = 1

center = (int(size[0]/2)+6, int(size[1]/2)-61)

img = np.zeros((size[0], size[1], 3), dtype=np.uint8)

img[:] = (0,0,0)

# draw point in center
img = cv2.circle(img, (center[1], center[0]), 0, (255, 255, 255), 2)

#for radius in range(step, int((size[0]**2 + size[1]**2)**(1/2)), step):
radius = step
while(radius < int((size[0]**2 + size[1]**2)**(1/2))):

  no_of_point = int(2*np.pi*radius / step)
  #print(f'Radius={radius} ', end='')

  for point in range(0, no_of_point):

    point_al = 2*np.pi/(no_of_point)*point
    y = center[1] + radius * math.sin(point_al)
    x = center[0] + radius * math.cos(point_al)

    y = int(y)
    x = int(x)
    img = cv2.circle(img, (y, x), 0, (255, 255, 255), thickness)


    #print(f'{point_al} ', end='')

  radius = radius + step/1.5 + radius/7
  print()
print(time.clock() - start_time, "seconds")
display(Image.fromarray(img))
ret = cv2.imwrite('/content/gdrive/MyDrive/circle.png', img, [cv2.IMWRITE_PNG_COMPRESSION, 0])
```
